```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Coffee Production

```python
total_production = pd.read_csv("./coffee_data/Coffee_production.csv")
```

```python
total_production.head(7)
```

```python
total_production.info()
```

```python
total_production[total_production.columns[2:]] = total_production[total_production.columns[2:]] * 0.001
total_production.head()
```

```python
year_production = total_production.drop(["Country", "Coffee type", "Total_production"], axis=1)
year_production.head()
```

```python
plt.figure(figsize=(13, 9))
sns.boxplot(year_production, orient="h")
plt.xlabel("Coffee Production (Tons)")
plt.ylabel("Years")
plt.title("Coffee Production Distribution From 1990-2020")
plt.show()
```

```python
# The function takes the total production dataframe to create a dataframe with coffee type production based on the countries
def coffee_type(total_production):
    coffee_t = total_production["Coffee type"].value_counts().index
    production_t = total_production["Coffee type"].value_counts().values
    
    coffeeT_production = pd.DataFrame({"Coffee Type": coffee_t, "Production": production_t})
    mix_value = coffeeT_production[coffeeT_production["Coffee Type"].str.contains("Arabica/Robusta|Robusta/Arabica")]["Production"].sum()
    coffeeT_production.loc[coffeeT_production["Coffee Type"] == "Arabica/Robusta", "Production"] = mix_value
    coffeeT_production = coffeeT_production.loc[coffeeT_production["Coffee Type"] != "Robusta/Arabica"]
    
    return coffeeT_production
```

```python
coffeeT_production = coffee_type(total_production)
```

```python
plt.figure(figsize=(8,7))
sns.barplot(x="Coffee Type", y="Production", data=coffeeT_production, order=["Arabica", "Arabica/Robusta", "Robusta"])
plt.ylabel("Amount of Countries")
plt.title("Most Produced Coffee Types")
plt.show()
```

```python
total_production.loc[total_production["Coffee type"] == "Robusta/Arabica", "Coffee type"] = "Arabica/Robusta"
```

```python
plt.figure(figsize=(7,8))
sns.barplot(x="Coffee type", y="Total_production", data=total_production, errorbar=None)
plt.ylabel("Total Production (Tons)")
plt.title("Total Production of Coffee Types from 1990-2020")
plt.show()
```

```python
top10_producers = total_production.sort_values(by="Total_production", ascending=False)[:10]
top10_producers
```

```python
plt.figure(figsize=(10,8))
sns.barplot(x="Country", y="Total_production", data=top10_producers, hue="Coffee type", dodge=False)
plt.title("Top 10 Total Coffee Producers")
plt.ylabel("Total Production (Tons)")
plt.show()
```

```python
coffee_prod_history = top10_producers.drop(["Coffee type", "Total_production"], axis=1).set_index("Country").T
```

```python
coffee_prod_history.head(6)
```

```python
plt.figure(figsize=(16, 9))
sns.lineplot(coffee_prod_history)
plt.xticks(rotation=45)
plt.grid(alpha=0.40)
plt.xlabel("Years")
plt.ylabel("Coffee Production (Tons)")
plt.title("Coffee Production History of Top 10 Producers")
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(coffee_prod_history.pct_change().corr(), annot=True)
plt.title("Correlation of coffee production yields of the top 10 coffee producers")
plt.show()
```

### Correlation

```python
production = total_production[["Country", "Coffee type"]]
production.iloc[1, 0] = "Bolivia"
production.head()
```

```python
precipitation = pd.read_csv("./cluster_tables/precipitation.csv")
```

```python
precipitation.info()
```

```python
precipitation.drop(["Rank", "Continent"], axis=1, inplace=True)
```

```python
precipitation.columns = ["Country", "Precipitation (mm)"]
precipitation.head()
```

```python
# remove comas and convert values to float
def convert_values(values):
    prec_num = []
    for value in values:
        c_value = ""
        for i in value:
            if i.isnumeric() == False:
                continue
            c_value += i
        prec_num.append(c_value)
    
    return prec_num
```

```python
precipitation["Precipitation (mm)"] = convert_values(precipitation["Precipitation (mm)"])
precipitation["Precipitation (mm)"] = precipitation["Precipitation (mm)"].astype(float)
```

```python
temperature = pd.read_csv("./cluster_tables/temperature.csv")
```

```python
temperature.info()
```

```python
temperature.columns = ["Country", "Temperature (Celsius)"]
temperature.head()
```

```python
production = production.join(precipitation.set_index("Country"), on="Country", 
                how="inner").join(temperature.set_index("Country"), on="Country", how="inner")
production.reset_index(drop=True, inplace=True)
```

```python
production["Temperature (Celsius)"] = production["Temperature (Celsius)"].astype(float)
production.set_index("Country", inplace=True)
```

```python
production.head()
```

```python
sns.heatmap(production.corr(numeric_only=True), annot=True)
plt.show()
```

They vary together but there is a weak association

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaled_val = scaler.fit_transform(production[["Precipitation (mm)", "Temperature (Celsius)"]])
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3, n_init="auto").fit(scaled_val)
```

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(x="Precipitation (mm)", y="Temperature (Celsius)", 
                hue=kmeans.labels_, data=production, palette="viridis", style="Coffee type")
plt.show()
```
