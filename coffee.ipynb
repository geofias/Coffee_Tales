```python
import pandas as pd
```

### Coffee Production

```python
total_production = pd.read_csv("./coffee_data/Coffee_production.csv")
```

```python
total_production.head(7)
```

```python
total_production.info()
```

```python
total_production[["Country", "Total_production"]].sort_values(by="Total_production", ascending=False)[:10]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
year_production = total_production.drop(["Country", "Coffee type", "Total_production"], axis=1)
year_production.head()
```

```python
plt.figure(figsize=(13, 9))
sns.boxplot(year_production, orient="h")
plt.xlabel("Coffee Production")
plt.ylabel("Years")
plt.title("Coffee Production From 1990-2020")
plt.show()
```

```python
plt.figure(figsize=(13, 8))
sns.lineplot(year_production.mean(), c="orange")
plt.grid(alpha=0.40)
plt.xticks(rotation=45)
plt.ylabel("Production Mean")
plt.title("Coffee Production From 1990-2020")
plt.show()
```

```python
# The function takes the total production dataframe to create a dataframe with coffee type production based on the countries
def coffee_type(total_production):
    coffee_t = total_production["Coffee type"].value_counts().index
    production_t = total_production["Coffee type"].value_counts().values
    
    coffeeT_production = pd.DataFrame({"Coffee Type": coffee_t, "Production": production_t})
    mix_value = coffeeT_production[coffeeT_production["Coffee Type"].str.contains("Arabica/Robusta|Robusta/Arabica")]["Production"].sum()
    coffeeT_production.loc[coffeeT_production["Coffee Type"] == "Arabica/Robusta", "Production"] = mix_value
    coffeeT_production = coffeeT_production.loc[coffeeT_production["Coffee Type"] != "Robusta/Arabica"]
    
    return coffeeT_production
```

```python
coffeeT_production = coffee_type(total_production)
```

```python
plt.figure(figsize=(8,7))
sns.barplot(x="Coffee Type", y="Production", data=coffeeT_production, order=["Arabica", "Arabica/Robusta", "Robusta"])
plt.ylabel("Amount of Countries")
plt.title("Coffee Production")
plt.show()
```

```python
total_production.loc[total_production["Coffee type"] == "Robusta/Arabica", "Coffee type"] = "Arabica/Robusta"
```

```python
plt.figure(figsize=(7,8))
sns.barplot(x="Coffee type", y="Total_production", data=total_production, errorbar=None)
plt.ylabel("Total Production")
plt.title("Total Production of Coffee Types from 1990-2020")
plt.show()
```

```python
top10_producers = total_production.sort_values(by="Total_production", ascending=False)[:10]
top10_producers
```

```python
plt.figure(figsize=(10,8))
sns.barplot(x="Country", y="Total_production", data=top10_producers, hue="Coffee type")
plt.title("Total Coffee Production")
plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.bar(x="Country", height="Total_production", data=top10_producers)
plt.show()
```

```python
coffee_prod_history = top10_producers.drop(["Coffee type", "Total_production"], axis=1).set_index("Country").T
```

```python
coffee_prod_history.head(6)
```

```python
plt.figure(figsize=(16, 9))
sns.lineplot(coffee_prod_history)
plt.xticks(rotation=45)
plt.grid(alpha=0.40)
plt.xlabel("Years")
plt.ylabel("Production")
plt.title("Coffee Production History of Top 10 Countries")
plt.show()
```

### Coffee Exportation

```python
coffee_export = pd.read_csv("coffee_data/Coffee_export.csv")
```

```python
coffee_export.head()
```

```python
coffee_export.info()
```

```python
year_export = coffee_export.drop(["Country", "Total_export"], axis=1)
year_export.head(6)
```

```python
plt.figure(figsize=(13, 9))
sns.boxplot(year_export, orient="h")
plt.xlabel("Coffee Exportation")
plt.ylabel("Years")
plt.title("Coffee Exportation From 1990-2019")
plt.show()
```

```python
coffee_export[coffee_export["2014"] < 0]
```

```python
plt.figure(figsize=(13, 8))
sns.lineplot(year_export.mean(), c="green")
plt.grid(alpha=0.40)
plt.xticks(rotation=45)
plt.ylabel("Export Mean")
plt.title("Coffee Exportation From 1990-2020")
plt.show()
```

```python
top10_exporters = coffee_export.sort_values(by="Total_export", ascending=False)[:10]
top10_exporters
```

```python
plt.figure(figsize=(10,8))
sns.barplot(x="Country", y="Total_export", data=top10_exporters)
plt.title("Total Coffee Exportation")
plt.show()
```

```python
coffee_exp_history = top10_exporters.drop(["Total_export"], axis=1).set_index("Country").T
coffee_exp_history.head(6)
```

```python
plt.figure(figsize=(16, 9))
sns.lineplot(coffee_exp_history)
plt.xticks(rotation=45)
plt.grid(alpha=0.40)
plt.xlabel("Years")
plt.ylabel("Exportation")
plt.title("Coffee Exportation History of Top 10 Countries")
plt.show()
```
